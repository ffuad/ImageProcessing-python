```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('F:/ACADEMIC/lena.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
im_height = img.shape[0]
im_width = img.shape[1]
print(im_height, im_width)
```

```python
import numpy as np

kernel = np.array(([0,-1,0],[-1,5,-1],[0,-1,0]), np.float32)
sizeKernel = 3
padding = sizeKernel//2;

imageWithBorder = cv2.copyMakeBorder(img, padding,padding,padding,padding,cv2.BORDER_REPLICATE)
im_height = imageWithBorder.shape[0]
im_width = imageWithBorder.shape[1]
print(imageWithBorder.shape)
```

```python
result = np.zeros((im_height, im_width), dtype="float32")

for x in range(im_height):
  for y in range(im_width):
    for i in range(sizeKernel):
      for j in range(sizeKernel):
        result[x][y] += imageWithBorder[x-i][y-j]*kernel[i][j]

print(result)
```

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
